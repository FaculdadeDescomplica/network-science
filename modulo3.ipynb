```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from string import ascii_uppercase
```

# Funções do Módulo 2

```python
def plot_degree2(g):
    degree_list = [degree for node, degree in g.degree]
    degree_series = pd.Series(degree_list)
    degree_series.hist()
    plt.xlabel('Grau')
    plt.ylabel('Frequência')
    plt.title('Distribuição dos graus')
```

```python
def plot_rank(g, log=False, weight=''):
    degree_list = [degree for node, degree in g.degree(weight=weight)]
    y = sorted(degree_list, reverse=True)
    x = range(len(y))
    plt.plot(x, y)
    plt.xlabel('Posição (rank)')
    plt.ylabel('Grau')
    if log:
        plt.loglog()
```

# 6. Exemplos com NetworkX

## Clube de karatê
Zachary, Wayne W. “An Information Flow Model for Conflict and Fission in Small Groups.” Journal of Anthropological Research, 33, 452–473, (1977).

```python
G = nx.karate_club_graph()
print(G)
```

```python
nx.draw_circular(G, with_labels=True)
```

### "Mundo pequeno"

```python
nx.density(G)
```

```python
nx.average_shortest_path_length(G)
```

```python
def get_distances(g):
    distance_list = []
    for source, dist in nx.shortest_path_length(g):
        distance_list += dist.values()
    return pd.Series(distance_list)
```

```python
distance_list = get_distances(G)
distance_list.describe()
```

### "Meus amigos são amigos"

```python
nx.average_clustering(G)
```

### "Desigualdade em redes"

```python
plot_degree2(G)
```

### "Tudo está conectado"

```python
nx.is_connected(G)
```
