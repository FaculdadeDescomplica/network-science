```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

# 2. Erdös-Rényi

Rede muda bastante de acordo com o parâmetro `p`, que é a probabilidade de uma aresta ser criada:

```python
n = 25
p = 0.01
G = nx.erdos_renyi_graph(n, p)
nx.draw_networkx(G, pos=nx.circular_layout(G))
```

```python
degree_list = [degree for node, degree in G.degree]
degree_list = pd.Series(degree_list)
degree_list.describe()
```

## Aumentando a rede

```python
n = 1000
p = 0.05
G = nx.erdos_renyi_graph(n, p, seed=42)
print(G)
```

## Densidade

```python
nx.density(G)
```

## Clusterização

```python
nx.average_clustering(G)
```

## Distância média

```python
nx.average_shortest_path_length(G)
```

## Distribuição dos graus

```python
degree_list = [degree for node, degree in G.degree]
degree_list = pd.Series(degree_list)
degree_list.hist();
```

# 3. Barabási-Albert

Rede vai depender do **número de nós `n`** e do **número de arestas a serem anexadas `m`**:

```python
n = 25
m = 2
G = nx.barabasi_albert_graph(n, m, seed=42)
nx.draw_networkx(G, pos=nx.circular_layout(G))
```

## Aumentando a rede

```python
n = 1000
m = 20
G = nx.barabasi_albert_graph(n, m, seed=42)
print(G)
```

## Densidade

```python
nx.density(G)
```

## Clusterização

```python
nx.average_clustering(G)
```

## Distância média

```python
nx.average_shortest_path_length(G)
```

## Distribuição dos graus

```python
degree_list = [degree for node, degree in G.degree]
degree_list = pd.Series(degree_list)
degree_list.hist();
```

```python
degree_list.describe()
```

### Usando o gráfico de *ranqueamento*

```python
y = degree_list.sort_values(ascending=False)
x = range(len(y))
plt.plot(x, y)
plt.loglog()
plt.xlabel('Posição (rank)')
plt.ylabel('Grau do nó');
```
