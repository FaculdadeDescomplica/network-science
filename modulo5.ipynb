```python
import networkx as nx
import matplotlib.pyplot as plt
```

# 2. Centralidade de Grau

## Rede não direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

Medida será normalizada pelo grau máximo possível:

```python
nx.degree_centrality(G)
```

## Rede direcionada (digrafo)

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

Soma das arestas de entrada e saída:

```python
nx.degree_centrality(H)
```

Considerando apenas **grau de entrada**:

```python
nx.in_degree_centrality(H)
```

Considerando apenas **grau de saída**:

```python
nx.out_degree_centrality(H)
```

# 3. Centralidade de Proximidade

## Rede não direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

```python
nx.closeness_centrality(G)
```

```python
G = nx.barbell_graph(5, 1)
nx.draw_networkx(G)
```

Nó "*do meio*" é mais central de acordo com a **proximidade**:

```python
nx.closeness_centrality(G)
```

## Rede direcionada (digrafo)

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

Por padrão, calcula a distância até **chegar** ao vértice:

```python
nx.closeness_centrality(H)
```

Para considerar a saída, precisamos usar `.reverse()`:

```python
nx.closeness_centrality(H.reverse())
```

# 4. Centralidade de Intermediação

## Rede não direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

Todos os caminhos mínimos passam pelo nó central:

```python
nx.betweenness_centrality(G)
```

```python
G = nx.barbell_graph(5, 1)
nx.draw_networkx(G)
```

Nenhum caminho mínimo passa pelos nós *agrupados*:

```python
nx.betweenness_centrality(G)
```

## Rede direcionada (digrafo)

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

```python
nx.betweenness_centrality(H)
```

# 5. Centralidade de Autovetor

## Rede não direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

Nó central "*empresta*" parte da sua importância para os demais:

```python
nx.eigenvector_centrality(G)
```

```python
G = nx.barbell_graph(5, 1)
nx.draw_networkx(G)
```

Maior centralidade para os "*nós líderes*" **4** e **6**:

```python
nx.eigenvector_centrality(G)
```

## Rede direcionada (digrafo)

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

Por padrão, valoriza as arestas de entrada:

```python
nx.eigenvector_centrality(H)
```

Para usar as arestas de saída, precisamos usar o `.reverse()`:

```python
nx.eigenvector_centrality(H.reverse())
```

# 6. Centralidade de PageRank

## Rede não direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

Assim como na centralidade de autovetor, o nó central "*empresta*" parte da sua importância para os demais:

```python
nx.pagerank(G)
```

Valores sempre vão somar **1**:

```python
sum(nx.pagerank(G).values())
```

```python
G = nx.barbell_graph(5, 1)
nx.draw_networkx(G)
```

```python
nx.pagerank(G)
```

## Rede direcionada (digrafo)

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

Por padrão, valoriza as arestas de entrada:

```python
nx.pagerank(H)
```

Para usar as arestas de saída, precisamos usar o `.reverse()`:

```python
nx.pagerank(H.reverse())
```
