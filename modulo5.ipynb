```python
import networkx as nx
import matplotlib.pyplot as plt
```

# 2. Centralidade de Grau

## Rede n√£o direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

Medida ser√° normalizada pelo grau m√°ximo poss√≠vel:

```python
nx.degree_centrality(G)
```

## Rede direcionada (digrafo)

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

Soma das arestas de entrada e sa√≠da:

```python
nx.degree_centrality(H)
```

Considerando apenas **grau de entrada**:

```python
nx.in_degree_centrality(H)
```

Considerando apenas **grau de sa√≠da**:

```python
nx.out_degree_centrality(H)
```

# 3. Centralidade de Proximidade

## Rede n√£o direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

```python
nx.closeness_centrality(G)
```

```python
G = nx.barbell_graph(5, 1)
nx.draw_networkx(G)
```

N√≥ "*do meio*" √© mais central de acordo com a **proximidade**:

```python
nx.closeness_centrality(G)
```

## Rede direcionada (digrafo)

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

Por padr√£o, calcula a dist√¢ncia at√© **chegar** ao v√©rtice:

```python
nx.closeness_centrality(H)
```

Para considerar a sa√≠da, precisamos usar `.reverse()`:

```python
nx.closeness_centrality(H.reverse())
```

# 4. Centralidade de Intermedia√ß√£o

```python
nx.betweenness_centrality(H)
```

# 5. Centralidade de Autovetor

## Rede n√£o direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

N√≥ central "*empresta*" parte da sua import√¢ncia para os demais:

```python
nx.eigenvector_centrality(G)
```

```python
G = nx.barbell_graph(5, 1)
nx.draw_networkx(G)
```

Maior centralidade para os "*n√≥s l√≠deres*" **4** e **6**:

```python
nx.eigenvector_centrality(G)
```

## Rede direcionada (digrafo)

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

Por padr√£o, valoriza as arestas de entrada:

```python
nx.eigenvector_centrality(H)
```

Para usar as arestas de sa√≠da, precisamos usar o `.reverse()`:

```python
nx.eigenvector_centrality(H.reverse())
```

# 6. Centralidade de PageRank

## Rede n√£o direcionada

```python
G = nx.star_graph(6)
nx.draw_networkx(G)
```

Assim como na centralidade de autovetor, o n√≥ central "*empresta*" parte da sua import√¢ncia para os demais:

```python
nx.pagerank(G)
```

Valores sempre v√£o somar **1**:

```python
sum(nx.pagerank(G).values())
```

```python
G = nx.barbell_graph(5, 1)
nx.draw_networkx(G)
```

```python
nx.pagerank(G)
```

## Rede direcionada (digrafo)

```python
H = nx.DiGraph({
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['D']
})
nx.draw_networkx(H)
```

Por padr√£o, valoriza as arestas de entrada:

```python
nx.pagerank(H)
```

Para usar as arestas de sa√≠da, precisamos usar o `.reverse()`:

```python
nx.pagerank(H.reverse())
```
